```python
dg=pd.read_csv("dentate_gyrus/human_rna_velocity_genes_dentate_gyrus.csv")
hnsc=pd.read_csv("hNSC/rank_velocity_genes_hNSC.csv", sep="\t").iloc[:, 1:]
nsp=pd.read_csv("neuronal_splicing/rank_velocity_genes_neuronal_splicing.csv", sep=",")
```

```python
#- all to upper case
dg = dg.apply(lambda x: x.astype(str).str.upper())
dg.head()
nsp=nsp.apply(lambda x: x.astype(str).str.upper())
nsp.head()
hnsc = hnsc.apply(lambda x: x.astype(str).str.upper())
hnsc.head()
```

```python
dg_arr=np.array(dg)
hnsc_arr=np.array(hnsc)
nsp_arr=np.array(nsp)
```

```python
nsp_set = set()
for l in nsp_arr:
    nsp_set.update(l)

hnsc_set = set()
for l in hnsc_arr:
    hnsc_set.update(l)

dg_set = set()
for l in dg_arr:
    dg_set.update(l)
```

```python
print(nsp_set & hnsc_set)
print(dg_set & hnsc_set)
print(nsp_set & dg_set)
print(nsp_set & dg_set & hnsc_set)
```
