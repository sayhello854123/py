```python
import numpy as np
import pandas as pd
```

```python
sampleList='/home/chelseax/data/IDC_patients.txt'
samples = np.loadtxt(sampleList, dtype='object')

k = -1
m = 0
for sp in samples: 
    for ele in ['exon','intron']:
        k += 1
        try:
            fpath='/pylon5/bi48n4p/chelseax/TCGA15/%s/%sCount.txt'%(sp, ele)
            tmp = pd.read_csv(fpath, sep='\t').iloc[:,[-1]]
            tmp.columns = ['cts']
            tmp['num'] = [i+1 for i in range(len(tmp.index))]
            tmp['ele'] = [ele] * len(tmp.index)
            tmp['sp'] = [sp] * len(tmp.index)
            df = tmp if k == 0 else df.append(tmp)
        except:
            m += 1
df['Histology'] = ['IDC'] * len(df.index)   
print(df.shape)
print(k, m)
df 
```

```python
df.to_csv('/pylon5/bi48n4p/chelseax/TCGA15/IDC_Cts.csv')
```

```python
import numpy as np
import pandas as pd

sampleList='/home/chelseax/data/ILC_patients.txt'
samples = np.loadtxt(sampleList, dtype='object')

k = -1
m = 0
for sp in samples: 
    for ele in ['exon','intron']:
        k += 1
        try:
            fpath='/pylon5/bi48n4p/chelseax/TCGA15/%s/%sCount.txt'%(sp, ele)
            tmp = pd.read_csv(fpath, sep='\t').iloc[:,[-1]]
            tmp.columns = ['cts']
            tmp['num'] = [i+1 for i in range(len(tmp.index))]
            tmp['ele'] = [ele] * len(tmp.index)
            tmp['sp'] = [sp] * len(tmp.index)
            df = tmp if k == 0 else df.append(tmp)
        except:
            m += 1
            print(sp)

df['Histology'] = ['ILC'] * len(df.index)            
print(df.shape)
print(k, m)
df 
```
